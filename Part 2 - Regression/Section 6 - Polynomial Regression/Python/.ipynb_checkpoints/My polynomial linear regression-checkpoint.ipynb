```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
#training the simple regression model on data set
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
#training the polynomial regression model on data set
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_req_2 = LinearRegression()
lin_req_2.fit(X_poly, Y)
```

```python
#visualising the linear regression
plt.scatter(X, Y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('truth or bluff')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#visualising the polynomial regression
plt.scatter(X, Y, color = 'red')
plt.plot(X, lin_req_2.predict(X_poly), color = 'blue')
plt.title('truth or bluff')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#predicting a new result with linear regression
lin_reg.predict([[6.5]])
```

```python
#predicting a new result with polynomial regression
lin_req_2 .predict(poly_reg.fit_transform([[6.5]]))
```
